```python
import json


with open("product_knowledge.json", "r", encoding="utf-8") as f:
    products = json.load(f)

chunks = []

for product in products:
    # دسته‌بندی
    category = product["category"].replace("|", " > ")

    # تبدیل رتبه و تعداد نظرات به عدد
    try:
        rating = float(product["rating"])
    except:
        rating = None

    try:
        rating_count = int(product["rating_count"].replace(",", ""))
    except:
        rating_count = None

    # تبدیل about_product به یک پاراگراف
    description = product["about_product"].replace("|", ". ")

    # استخراج عناوین نقدها (ممکنه در چند رکورد تکرار شده باشن، پس ادغام می‌کنیم)
    review_titles = []
    for review_group in product["reviews"]:
        titles = review_group.get("review_title", "")
        if isinstance(titles, str):
            review_titles.extend([title.strip() for title in titles.split(",") if title.strip()])

    # حذف تکراری‌ها و چسباندن
    unique_titles = list(set(review_titles))
    reviews = " | ".join(unique_titles)

    # ساخت چانک
    chunk = {
        "id": product["product_id"],
        "name": product["product_name"],
        "description": description,
        "metadata": {
            "product_id": product["product_id"],
            "category": category,
            "rating": rating,
            "rating_count": rating_count,
            "reviews": reviews
        }
    }

    chunks.append(chunk)

# ذخیره چانک‌ها
with open("product_chunks_cleaned.json", "w", encoding="utf-8") as f:
    json.dump(chunks, f, ensure_ascii=False, indent=2)

# مثال دیدن یک چانک
print(json.dumps(chunks[0], indent=2, ensure_ascii=False))
```

```python
from sentence_transformers import SentenceTransformer
import json
from tqdm import tqdm

# مدل سبک و دقیق برای امبدینگ
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# بارگذاری داده‌های آماده شده (چانک‌ها)
with open("product_chunks_cleaned.json", "r", encoding="utf-8") as f:
    data = json.load(f)

embeddings_data = []

for item in tqdm(data):
    name = item.get("name", "")
    description = item.get("description", "")  # اطمینان از وجود description
    reviews = item.get("metadata", {}).get("reviews", "")

    text = f"{name}. {description}. Reviews: {reviews}"

    # گرفتن بردار امبدینگ
    embedding = embedding_model.encode(text)

    embeddings_data.append({
        "embedding": embedding.tolist(),
        "metadata": {
            "id": item.get("id", ""),
            "name": name,
            "category": item.get("metadata", {}).get("category", ""),
            "rating": item.get("metadata", {}).get("rating", None),
            "rating_count": item.get("metadata", {}).get("rating_count", None),
            "reviews": reviews,
            "description": description  # اضافه کردن description به metadata
        }
    })

# ذخیره بردارها در فایل JSON
with open("product_embeddings.json", "w", encoding="utf-8") as f:
    json.dump(embeddings_data, f, ensure_ascii=False, indent=2)

print("Embedding generation and saving completed!")

# چک کردن اولین مورد
import pprint
with open("product_embeddings.json", "r", encoding="utf-8") as f:
    embeddings = json.load(f)
print(f"Total embeddings: {len(embeddings)}")
pprint.pprint(embeddings[0])
```

تابع سیمیلرتی میسازیم که نزدیک ترین داده نسبت به کیوری را پیدا کند 

```python
def search_similar_products(query_text, embedding_model, index, embeddings_data, top_k=5):
    query_embedding = embedding_model.encode(query_text).astype('float32')
    D, I = index.search(np.array([query_embedding]), k=top_k)
    results = []
    for rank, idx in enumerate(I[0]):
        dist = D[0][rank]
        item = embeddings_data[idx]
        metadata = item["metadata"]
        results.append({
            "rank": rank + 1,
            "name": metadata.get("name", ""),
            "description": metadata.get("description", ""),  # اضافه شد
            "category": metadata.get("category", ""),
            "rating": metadata.get("rating", ""),
            "rating_count": metadata.get("rating_count", ""),
            "distance": round(float(dist), 4),
            "metadata": metadata
        })
    return results
```

```python
query = "I need a rechargeable AAA battery with long-lasting power"
results = search_similar_products(query, embedding_model, index, embeddings_data)

for r in results:
    print(f"Rank {r['rank']}:")
    print(f"  Name: {r['name']}")
    print(f"  Description: {r['description']}")  # اضافه شد
    print(f"  Category: {r['category']}")
    print(f"  Rating: {r['rating']} ({r['rating_count']} votes)")
    print(f"  Distance: {r['distance']}")
    print("-" * 40)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# بارگذاری مدل TinyLlama
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float32,  # چون فقط CPU داریم
    device_map="cpu"  # اجرا روی CPU
)
```

```python
def build_prompt_cpu(query, context_list, max_desc_len=200, max_review_len=150):
    """
    ساخت پرامپت کوتاه برای CPU با محدودیت طول متن.
    """
    context_texts = []
    for doc in context_list[:3]:  # فقط 3 محصول برتر
        # کوتاه کردن description
        desc = doc['description'][:max_desc_len] + ("..." if len(doc['description']) > max_desc_len else "")
        # کوتاه کردن reviews
        reviews = doc['metadata'].get('reviews', "")[:max_review_len] + ("..." if len(doc['metadata'].get('reviews', "")) > max_review_len else "")
        context_texts.append(f"- {doc['name']}: {desc} Reviews: {reviews} (Rating: {doc['rating']}, {doc['rating_count']} votes)")

    context_text = "\n".join(context_texts)
    prompt = f"""You are a helpful product recommendation assistant. Answer in 1-2 sentences using the product info below. Be concise.

Question: {query}

Products:
{context_text}

[Answer]: """
    return prompt

def generate_answer_cpu(model, tokenizer, prompt, max_new_tokens=150):
    """
    تولید پاسخ روی CPU با تنظیمات بهینه.
    """
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512)
    inputs = {k: v.to("cpu") for k, v in inputs.items()}
    outputs = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        num_return_sequences=1,
        do_sample=True,
        top_k=40,
        top_p=0.9,
        temperature=0.5,
        no_repeat_ngram_size=3,
        pad_token_id=tokenizer.eos_token_id
    )
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    # فقط بخش بعد از [Answer]:
    if "[Answer]:" in response:
        response = response.split("[Answer]:")[1].strip()
    else:
        response = response.strip()
    return response

def rag_pipeline_cpu(query, top_k=5):
    """
    Pipeline RAG برای CPU
    """
    # 1. جستجوی محصولات مشابه
    context_list = search_similar_products(query, embedding_model, index, embeddings_data, top_k=top_k)
    # 2. ساخت پرامپت کوتاه
    prompt = build_prompt_cpu(query, context_list)
    # 3. تولید پاسخ
    response = generate_answer_cpu(model, tokenizer, prompt)
    return response, context_list
```

```python
query = "I need a rechargeable AAA battery with long-lasting power"
response, retrieved_docs = rag_pipeline_cpu(query)

print("🔎 پرس‌وجو:", query)
print("\n📚 محصولات بازیابی‌شده:")
for doc in retrieved_docs:
    print(f"Rank {doc['rank']}: {doc['name']} (Rating: {doc['rating']}, Distance: {doc['distance']})")
    print(f"Description: {doc['metadata'].get('description', 'No description')}")
print("\n📝 پاسخ نهایی:\n", response)
```

اموزش مدل

```python
import os
import json
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, Trainer, TrainingArguments
from datasets import load_dataset
from peft import LoraConfig, get_peft_model, TaskType

# ---------------------------
# مسیرها
# ---------------------------
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
data_file = "./calls_formatted_lora.jsonl"
model_dir = "./tinyllama_finetuned"
log_dir = "./logs/tinyllama"

# ایجاد دایرکتوری‌ها
os.makedirs(model_dir, exist_ok=True)
os.makedirs(log_dir, exist_ok=True)

# ---------------------------
# بارگذاری مدل و tokenizer
# ---------------------------
print("Loading model and tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(model_name)

# اضافه کردن pad_token اگر وجود ندارد
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    device_map="auto" if torch.cuda.is_available() else "cpu",
    trust_remote_code=True
)

print(f"Model loaded on device: {model.device}")

# ---------------------------
# تنظیم LoRA
# ---------------------------
print("Setting up LoRA configuration...")

# پیدا کردن target modules مناسب برای TinyLlama
target_modules = []
for name, module in model.named_modules():
    if "q_proj" in name or "v_proj" in name or "k_proj" in name or "o_proj" in name:
        target_modules.append(name.split(".")[-1])

# حذف تکراری‌ها
target_modules = list(set(target_modules))
print(f"Target modules found: {target_modules}")

# اگر target modules پیدا نشد، از نام‌های عمومی استفاده کنیم
if not target_modules:
    target_modules = ["q_proj", "v_proj", "k_proj", "o_proj", "gate_proj", "up_proj", "down_proj"]
    print(f"Using default target modules: {target_modules}")

lora_config = LoraConfig(
    r=16,  # rank افزایش یافت برای بهتر شدن کیفیت
    lora_alpha=32,  # alpha هم افزایش یافت
    target_modules=target_modules,
    lora_dropout=0.1,
    bias="none",
    task_type=TaskType.CAUSAL_LM
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()

# ---------------------------
# بارگذاری دیتاست و توکنایز
# ---------------------------
print("Loading and processing dataset...")
dataset = load_dataset("json", data_files={"train": data_file})["train"]

def tokenize_for_training(examples):
    """بهتر شده tokenization function"""
    # ترکیب prompt و completion
    texts = []
    for prompt, completion in zip(examples["prompt"], examples["completion"]):
        full_text = prompt + completion
        texts.append(full_text)
    
    # توکنایز کردن
    tokenized = tokenizer(
        texts,
        truncation=True,
        max_length=512,
        padding=False,  # padding را در dataloader انجام می‌دهیم
        return_tensors=None
    )
    
    # labels برابر input_ids
    tokenized["labels"] = tokenized["input_ids"].copy()
    
    return tokenized

# اعمال tokenization
tokenized_dataset = dataset.map(
    tokenize_for_training,
    batched=True,
    remove_columns=dataset.column_names,
    desc="Tokenizing dataset"
)

print(f"Dataset size: {len(tokenized_dataset)}")

# ---------------------------
# Data Collator برای padding
# ---------------------------
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False,  # برای causal LM
    pad_to_multiple_of=8 if torch.cuda.is_available() else None
)

# ---------------------------
# تنظیم TrainingArguments
# ---------------------------
# بررسی نسخه transformers برای تطبیق پارامترها
import transformers
print(f"Transformers version: {transformers.__version__}")

# تنظیمات اصلی
training_args_dict = {
    "output_dir": model_dir,
    "logging_dir": log_dir,
    
    # Training parameters
    "num_train_epochs": 3,  # افزایش epochs
    "per_device_train_batch_size": 2 if torch.cuda.is_available() else 1,
    "gradient_accumulation_steps": 4,  # برای شبیه‌سازی batch size بزرگتر
    
    # Optimization
    "learning_rate": 2e-4,
    "weight_decay": 0.01,
    "warmup_steps": 100,
    
    # Logging and saving
    "logging_steps": 10,
    "save_steps": 100,
    "save_total_limit": 3,
    
    # Other settings
    "dataloader_drop_last": False,
    "fp16": torch.cuda.is_available(),  # فقط با GPU
    "bf16": False,
    "remove_unused_columns": False,
    "report_to": [],  # هیچ reporting
    "load_best_model_at_end": False,
    "overwrite_output_dir": True,
    
    # Memory optimization
    "dataloader_pin_memory": False,
    "group_by_length": True,  # برای کارایی بهتر
}

# اضافه کردن evaluation_strategy بر اساس نسخه
try:
    # تست کردن کدام پارامتر کار می‌کند
    test_args = TrainingArguments(output_dir="./test", evaluation_strategy="no")
    training_args_dict["evaluation_strategy"] = "no"
    print("Using 'evaluation_strategy' parameter")
except TypeError:
    try:
        test_args = TrainingArguments(output_dir="./test", eval_strategy="no") 
        training_args_dict["eval_strategy"] = "no"
        print("Using 'eval_strategy' parameter")
    except TypeError:
        # اگر هیچکدام کار نکرد، بدون evaluation strategy ادامه دهیم
        print("No evaluation strategy parameter found")

training_args = TrainingArguments(**training_args_dict)

print(f"Training arguments configured:")
print(f"  - Epochs: {training_args.num_train_epochs}")
print(f"  - Batch size: {training_args.per_device_train_batch_size}")
print(f"  - Learning rate: {training_args.learning_rate}")
print(f"  - FP16: {training_args.fp16}")

# ---------------------------
# Trainer و آموزش
# ---------------------------
print("Initializing trainer...")
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator,
    tokenizer=tokenizer
)

# شروع آموزش
print("Starting training...")
try:
    trainer.train()
    print("✅ Training completed successfully!")
    
    # ذخیره مدل و tokenizer
    print("Saving model and tokenizer...")
    trainer.save_model()
    tokenizer.save_pretrained(model_dir)
    
    print(f"✅ Model and tokenizer saved to: {model_dir}")
    
    # نمایش فایل‌های ذخیره شده
    saved_files = os.listdir(model_dir)
    print("Saved files:")
    for file in saved_files:
        print(f"  - {file}")
        
except Exception as e:
    print(f"❌ Training failed with error: {e}")
    raise e
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel
import torch, os

checkpoint = "./tinyllama_finetuned/checkpoint-15"

print("Files in checkpoint-15:", os.listdir(checkpoint))

# بارگذاری tokenizer
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# بارگذاری base
base = AutoModelForCausalLM.from_pretrained(
    checkpoint,
    torch_dtype=torch.float32,
    trust_remote_code=True,
    low_cpu_mem_usage=True
).to("cpu")
base.eval()
print("✅ Base loaded on", next(base.parameters()).device)

# اعمال adapter
model_with_adapter = PeftModel.from_pretrained(base, checkpoint).to("cpu")
model_with_adapter.eval()
print("✅ Adapter applied successfully. Model ready.")

device = torch.device("cpu")   # یا "cuda" اگر کارت داری
```

```python
query = "Hi, this is Ali. My order #ORD12345 hasn't arrived and I want a refund."
inputs = tokenizer(query, return_tensors="pt", truncation=True, padding=True).to(device)

with torch.no_grad():
    out = model_with_adapter.generate(
        **inputs,
        max_new_tokens=200,
        do_sample=False,
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=tokenizer.eos_token_id
    )

print("=== OUTPUT ===")
print(tokenizer.decode(out[0], skip_special_tokens=True))
```

```python
# === Cell: run a single extraction prompt and print JSON part ===
import torch, re, json

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

prompt = (
    "You are a call-analysis AI. Given the transcript below, extract these fields as JSON: "
    "type, emotion, product_id, order_id, customer_name. "
    "Output ONLY valid JSON, no explanations, no extra text, no code.\n\n"
    "Transcript:\n"
    "Hi, this is Ali. My order #ORD12345 hasn't arrived and I want a refund.\n\n"
    "JSON:"
)

# توکنایز و بردن به device مدل
inputs = tokenizer(prompt, return_tensors="pt", truncation=True, padding=True).to(device)

with torch.no_grad():
    out = model_with_adapter.generate(
        **inputs,
        max_new_tokens=200,
        do_sample=False,               # deterministic
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=tokenizer.eos_token_id
    )

text = tokenizer.decode(out[0], skip_special_tokens=True)

print("=== RAW MODEL OUTPUT ===")
print(text)

# --- استخراج اولین JSON معتبر از خروجی ---
matches = re.findall(r"\{.*?\}", text, flags=re.DOTALL)
json_output = None
if matches:
    candidate = matches[0]
    try:
        json_output = json.loads(candidate)
    except json.JSONDecodeError:
        pass

print("\n=== CLEAN JSON ===")
if json_output:
    print(json_output)
else:
    print("⚠️ No valid JSON found")
```

```python
#ایمپورت‌ها
from transformers import AutoTokenizer
from peft import PeftModel
from sentence_transformers import SentenceTransformer
import torch
import numpy as np
import faiss
import json
```

```python
# ---------------------------
# بارگذاری مدل فاین‌تیون شده
# ---------------------------
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
import torch

# مسیر آداپتر ذخیره شده
model_dir = "./my_finetuned_model"

# همون مدل پایه‌ای که برای آموزش استفاده کردم
base_model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

# بارگذاری توکنایزر
tokenizer = AutoTokenizer.from_pretrained(base_model_name)

# بارگذاری مدل پایه
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_name,
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    device_map="auto" if torch.cuda.is_available() else "cpu"
)

# سوار کردن آداپتر روی مدل پایه
model = PeftModel.from_pretrained(base_model, model_dir)
model.eval()

print("✅ Finetuned model loaded successfully!")
```

```python
#بارگذاری امبدینگ و ایندکس FAISS
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

with open("product_embeddings.json", "r", encoding="utf-8") as f:
    embeddings_data = json.load(f)

vectors = np.array([item["embedding"] for item in embeddings_data]).astype('float32')
index = faiss.IndexFlatL2(vectors.shape[1])
index.add(vectors)
```

```python
#تابع جستجوی محصولات مشابه
def search_similar_products(query_text, embedding_model, index, embeddings_data, top_k=5):
    query_embedding = embedding_model.encode(query_text).astype('float32')
    D, I = index.search(np.array([query_embedding]), k=top_k)
    results = []
    for rank, idx in enumerate(I[0]):
        dist = D[0][rank]
        item = embeddings_data[idx]
        metadata = item["metadata"]
        results.append({
            "rank": rank + 1,
            "name": metadata.get("name", ""),
            "description": metadata.get("description", ""),
            "category": metadata.get("category", ""),
            "rating": metadata.get("rating", ""),
            "rating_count": metadata.get("rating_count", ""),
            "distance": round(float(dist), 4),
            "metadata": metadata
        })
    return results
```

```python
def build_finetuned_prompt(query, context_list, max_products=3, max_desc_len=50):
    product_lines = []
    for doc in context_list[:max_products]:
        desc = (doc.get("description", "") or "")[:max_desc_len].strip()
        rating = doc.get("rating", "N/A")
        product_lines.append(f"- {doc['name']} | Rating: {rating}")
    products = "\n".join(product_lines)
    prompt = f"""Task: Recommend ONE product in exact format.
Example 1:
User: "I want wireless headphones for gym"
Products: - Sony WH-1000XM4 | Rating: 4.5
Response:
JSON: {{"emotion": "excited", "recommended_product": "Sony WH-1000XM4", "reason": "Great for gym with noise cancellation"}}
Text: Sony WH-1000XM4 is perfect for your gym sessions with excellent sound quality.
Example 2:
User: "Need waterproof speaker for beach"
Products: - JBL Clip 3 | Rating: 4.2
Response:
JSON: {{"emotion": "helpful", "recommended_product": "JBL Clip 3", "reason": "Waterproof and portable for beach use"}}
Text: JBL Clip 3 offers waterproof protection perfect for beach activities.
Now your turn:
User: "{query}"
Products:
{products}
Response:"""
    return prompt
```

```python
def generate_answer(model, tokenizer, prompt):
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512)
    
    # تعریف stop sequences
    stop_sequences = ["Example 3:", "User:", "\nUser:", "Example", "\nExample"]
    
    with torch.no_grad():
        outputs = model.generate(
            inputs.input_ids,
            max_new_tokens=120,
            do_sample=True,
            temperature=0.3,
            top_p=0.9,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
            early_stopping=True,
            repetition_penalty=1.1
        )
    
    response = tokenizer.decode(outputs[0][inputs.input_ids.shape[1]:], skip_special_tokens=True)
    
    # Manual stop sequence checking
    for stop_seq in stop_sequences:
        if stop_seq in response:
            response = response.split(stop_seq)[0]
            break
    
    return response.strip()
```

```python
# تست
test_query = "I need a Bluetooth speaker with waterproof protection"

# اینجا همه آرگومان‌ها رو بدیم
context_list = search_similar_products(
    test_query, 
    embedding_model, 
    index, 
    embeddings_data, 
    top_k=5
)

# ساخت پرامپت با کانتکست
prompt = build_finetuned_prompt(test_query, context_list)
# گرفتن خروجی مدل
response = generate_answer(model, tokenizer, prompt)

# نمایش نتایج
print("🔎 Query:", test_query)
print("\n📚 Retrieved Products:")
for doc in context_list:
    print(f"Rank {doc['rank']}: {doc['name']} (Rating: {doc['rating']}, Distance: {doc['distance']})")

print("\n📝 Model Response:\n", response)
```
